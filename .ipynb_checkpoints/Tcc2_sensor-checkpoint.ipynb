```julia
#= Vamos assumir uma tubulação com distancia 100 e diametro 5

                                    Distancia entre os transdutores
                     |----------------------------------------------------|
                     .                                                    .
                     .                                                    .
                     .                                                    .            
                     .                                                    .                                                    
                     .                        100                         .
|--------------------T-----------------------------------------------------------------------------------|
| -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> ->> |
| -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> ->> | 5
| -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> -> ->> |
|-------------------------------------------------------------------------T------------------------------|

Dessa forma é possivel distribuir os transdutores no espaço entre 0 e 100 pontos da tubulção.

vamos assumir que cada ponto dos 100 é igual a 1cm, logo temos uma tubulação de 1 metro.

Dentro dessa tubulação está passando um fluido, no caso deste exemplo, agua.

Neste exemplo iremos assumir que o fluido varia de forma linear, e iremos calcular o fluxo em cada passo baseado nas funções acima.

Nota: Neste exemplo será ignorado alguns fatores de fluidos, como a questão de fluxo laminar, gradiente de pressão, e perfil de velocidade parabolico




A ideia principal é:

* Criar uma lista de velocidades linear
    [0.1,0.2,0.3,...,9.9,10.0]
* Criar uma lista de distancia entre os transdutores (Nesse caso será feito uma progressão linear da distancia)
    [0.1,0.2,0.3....99.9,100.00]
* Calcular o Angulo e a distancia real percorrida pela onda

                                     |----------|
                                     | Calcular |
  [Distancia entre transdutor]  ---> |  Angulo  |  ---> [Alpha, Distancia_L]
                                     |----------|

* Calcular o tempo que a onda ultrassonica leva para percorrer a Distancia_L


                                         |----------|
                                         | Calcular |
  [Alpha, Distancia_L, Velocidade]  ---> |  Tempo   |  ---> [Tempo_subida, Tempo_Descida]
                                         |----------|

* Aplicar ruidos de leitura nos dados de tempo

* Calcular a velocidade baseada nos tempos de subida e descida


                                                          |----------|
                                                          | Calcular |
  [Alpha, Distancia_L, Tempo_subida, Tempo_descida]  ---> |  Tempo   |  ---> [Velocidade]
                                                          |----------|

* Plotar grafico de velocidade real x velocidade calculada

* Calcular e plotar curva de erro

* Criar tabela relacionando distancia entre transdutores e valor calculado

=#


```

```julia
using Plots
import Pkg; Pkg.add("DataFrames")
```

```julia
function calc_angle(transduct_dis::Float64, diameter::Float64)
    theta_rad::Float64 = atan(diameter / transduct_dis)
    theta_degree::Float64 = rad2deg(theta_rad)
    L_dis::Float64 = sqrt(transduct_dis^2 + diameter^2)
    return [theta_degree, L_dis]
end
```

```julia
function calc_time(angle::Float64, L_dis::Float64, vel::Float64)
    c::Float64 = 1480.00
    time_down = L_dis / (c + vel * cosd(angle))
    time_up = L_dis / (c - vel * cosd(angle))
    return [time_up, time_down, vel, angle, L_dis]
end
```

```julia
function calc_velocity(angle::Float64, L_dis::Float64, tu::Float64, td::Float64)
    delta_time = tu - td
    v::Float64 = (L_dis * delta_time) / (2 * cosd(angle) * tu * td)
    return v
end
```

```julia
function add_noise(value::Float64, noise_level::Float64)
    return value + (randn()*2-1) * noise_level
end
```

```julia
# Função para calcular o perfil de velocidade parabólico
function parabolic_velocity_profile(r, R, vmax)
    return vmax * (1 - (r / R)^2)
end
```

```julia
# Função para calcular o tempo de trânsito usando o perfil parabólico
function calc_time_parabolic(angle, L_dis, R, vmax)
    c = 1480.0  # velocidade do som na água em m/s
    
    # Integrar o perfil de velocidade para obter a velocidade média ao longo do diâmetro do tubo
    num_points = 100  # Número de pontos para a integração
    rs = range(-R, stop=R, length=num_points)
    vs = parabolic_velocity_profile.(rs, R, vmax)
    
    avg_velocity = mean(vs)
    
    time_down = L_dis / (c + avg_velocity * cosd(angle))
    time_up = L_dis / (c - avg_velocity * cosd(angle))
    
    return time_up, time_down, avg_velocity, angle, L_dis
end
```

```julia
velocitys = range(0.1, 10.0, length=50)
distance_btw_transductores = range(0.1, 100.0, length=50)  # cm
diameters = range(0.1, 10.0, length=10)  # cm
results = []
noise_level_time = 1e-6  # Nível de ruído para o tempo (segundos)
noise_level_velocity = 0.01  # Nível de ruído para a velocidade (m/s)
```

```julia
# Calcular os ângulos e distâncias para diferentes distâncias entre transdutores e diâmetros
angle_L_dis = []
for transduct_dis in distance_btw_transductores
    for diameter in diameters
        push!(angle_L_dis, (transduct_dis, diameter, calc_angle(transduct_dis, diameter)))
    end
end
```

```julia
# Calcular os tempos e as velocidades para cada combinação de parâmetros
times = []
for vel in velocitys
    for (transduct_dis, diameter, (angle, L_dis)) in angle_L_dis
        push!(times, (transduct_dis, diameter, calc_time(angle, L_dis, vel)))
    end
end
```

```julia
velocity_measured = []
for (transduct_dis, diameter, (tu, td, vel, angle, L_dis)) in times
    v_measured = calc_velocity(angle, L_dis, tu, td)
    # Adiciona ruído à velocidade medida
    v_measured_noisy = add_noise(v_measured, noise_level_velocity)
    
    push!(velocity_measured, (transduct_dis, diameter, vel, v_measured_noisy))
end
```

```julia
df = DataFrame(transduct_dis = Float64[], diameter = Float64[], V_in = Float64[], V_out = Float64[], Error = Float64[])
for (transduct_dis, diameter, V_in, V_out) in velocity_measured
    error = V_out - V_in
    push!(df, (transduct_dis, diameter, V_in, V_out, error))
end
```

```julia
# Converter as unidades para cm e m/s
df.transduct_dis .= df.transduct_dis ./ 100  # Convertendo para metros
df.V_in .= df.V_in  # Já está em m/s
df.V_out .= df.V_out  # Já está em m/s
print("done")
```

```julia
# Plotar a velocidade real vs. velocidade calculada
vel_real = df.V_in
vel_calc = df.V_out
plot(vel_real, vel_calc, xlabel="Velocidade Real (m/s)", ylabel="Velocidade Calculada (m/s)", title="Velocidade Real vs. Velocidade Calculada", label="Velocidade Calculada", legend=:topleft)
plot!(x -> x, label="Ideal", linestyle=:dash)  # Linha ideal y=x
```
